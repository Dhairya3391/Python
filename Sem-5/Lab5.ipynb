### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
def min_max_scaling(age):
    min_age = df['Age'].min()
    max_age = df['Age'].max()
    return (age - min_age) / (max_age - min_age)
def decimal_scaling(age):
    max_abs_age = df['Age'].abs().max()
    return age / (10 ** len(str(int(max_abs_age))))
def z_score_scaling(age):
    mean_age = df['Age'].mean()
    std_age = df['Age'].std()
    return (age - mean_age) / std_age

df['Age_MinMax'] = df['Age'].apply(min_max_scaling)
df['Age_Decimal'] = df['Age'].apply(decimal_scaling)
df['Age_ZScore'] = df['Age'].apply(z_score_scaling)

df[['Age', 'Age_MinMax', 'Age_Decimal', 'Age_ZScore']].head()
```
