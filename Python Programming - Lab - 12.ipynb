### 04) WAP to demonstrate the use of Pie chart.

```python
x = ['A', 'B', 'C', 'D']
y = [20, 30, 10, 40]

plt.pie(y, labels=x, autopct='%1.1f%%')
plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

```python
# WAP to demonstrate the use of Bar chart.

x = ['A', 'B', 'C', 'D']
y = [20, 30, 10, 40]

plt.bar(x, y)
plt.show()
```

### 06) WAP to demonstrate the use of Scatter Plot.

```python
# WAP to demonstrate the use of Scatter plot.
import random

x = [random.randint(1, 100) for i in range(1, 51)]
y = [random.randint(1, 100) for i in range(1, 51)]

plt.scatter(x, y)
plt.show()
```

### 07) WAP to demonstrate the use of Histogram.

```python
# WAP to demonstrate the use of Histogram.

x = [random.randint(1, 100) for i in range(1, 51)]

plt.hist(x, bins=10)
plt.show()
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = ['A', 'B', 'C', 'D']
y = [20, 30, 10, 40]

for i in range(len(x)):
    plt.text(i, y[i], y[i], ha='center', va='bottom')
    
plt.bar(x, y)
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = [random.randint(1, 100) for i in range(1, 51)]
y = [random.randint(1, 100) for i in range(1, 51)]

colors = [random.randint(1, 100) for i in range(1, 51)]

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```
